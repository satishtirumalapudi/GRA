```python
def common_tags(i_tags, j_tags):
    i_tags_set = set(i_tags)
    j_tags_set = set(j_tags)
 
    if (i_tags_set & j_tags_set):
        c_tags = list(i_tags_set & j_tags_set)
    else:
        c_tags = None
        
    return c_tags
    
res = {}
c = 1
for i in range(len(keys)):
    i_tags = sample2[keys[i]][1]

    for j in range(i+1, len(keys)):
        e_tags = []
        
        j_tags = sample2[keys[j]][1]
        c_tags = common_tags(i_tags, j_tags)
        
        if c_tags:
            e_tags.append(c_tags)
            e_tags.append([keys[i], keys[j]])
            wts = []
            
            for tag in c_tags:
                # for i_tag
                i_index = sample2[keys[i]][1].index(tag) 
                i_wt = sample2[keys[i]][0][i_index]
                
                # for j_tag
                j_index = sample2[keys[j]][1].index(tag) 
                j_wt = sample2[keys[j]][0][j_index]
                
                wts.append([i_wt, j_wt])
            
            e_tags.append(wts)
            
        res["e{}".format(c)] = e_tags
        c+=1
```
